```python
post_login_url = 'https://sdsu.metabim.com/Default.htm'
#f = open(r"C:\Users\kwsmith\Desktop\this_file.txt")
#pkey = f.read()
#f.close()
uname = input("Enter FIS Username:")
pkey = getpass.getpass(prompt="Please enter FIS Password:",)
payload = {'ukey':uname, 'pass':pkey}
```

```python
#Uncomment the Appropriate Table

#Buildings:
table_url = 'https://sdsu.metabim.com/bldg/Buildings.csv?Qabc=K'
#ADA
#table_url = 'https://sdsu.metabim.com/access/Access.csv'
#Space Inventory
#table_url = 'https://sdsu.metabim.com/room/Rooms.csv?Qoper=EQ'
#Grounds
#table_url = 'https://sdsu.metabim.com/grnd/Grounds.csv'
#Departments
#table_url ='https://sdsu.metabim.com/dept/Departments.csv'
#Equipment
#table_url = 'https://sdsu.metabim.com/eqpt/Equipment.csv'
#Contacts
#table_url = 'https://sdsu.metabim.com/names/Contacts.csv?Qabc=B'
#Conditions
#table_url = 'https://sdsu.metabim.com/eqpt/Equipment.csv?Tkey=2'
#Users
#table_url = 'https://sdsu.metabim.com/admin/users/Accounts.csv?Qabc=!&Qstatus=A'
```

```python
with requests.Session() as session:
    post = session.post(post_login_url, data=payload)
    download = session.get(table_url)
    decoded_content = download.content.decode('utf-8')

#This converts the response text to a csv list    
cr = csv.reader(decoded_content.splitlines(), delimiter=',')
my_list = list(cr)

#This converts the list to a Pandas dataFrame
df = pd.DataFrame.from_records(my_list)
df.columns= df.iloc[0]
df.drop(index=0, axis=0, inplace=True)
df
```
